```python
def remove(df,name):
    iqr =1.5*(np.percentile(df[name],75)- np.percentile(df[name], 25))
    df.drop(df[df[name]> (iqr+ np.percentile(df[name],75))].index, inplace = True)
    df.drop(df[df[name] < (np.percentile(df[name],25)- iqr)].index, inplace = True)
    
remove(df,'Perimeter')
```

```python
#Dividing the independent and dependent features
X = df.iloc[:, : -1]

Y = df.iloc[: ,-1]
```

```python
print('Independent Feature Set Shape : ', X.shape)
print('Dependent Feature Set Shape   : ', Y.shape)
```

```python
#Splitting the data into train and test splits
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 4)
```

```python
print('Training data shape   : ', x_train.shape)
print('Training labels shape : ', y_train.shape)
print('Testing data shape    : ', x_test.shape)
print('Testing labels shape  : ', y_test.shape)
```

```python
x_train
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_test[['Area', 'Perimeter','AspectRation']] = scaler.fit_transform(x_test[['Area', 'Perimeter','AspectRation']])
x_test
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train[['Area', 'Perimeter','AspectRation']] = scaler.fit_transform(x_train[['Area', 'Perimeter','AspectRation']])
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 10, metric = 'minkowski', p = 2)
classifier.fit(x_train, y_train)
```

```python
print('Training Set Score : ', classifier.score(x_train, y_train))
print('Testing Set Score  : ', classifier.score(x_test, y_test))
```

```python
acc_train = []
acc_test  = []

for k in range(1,21):
    classifier = KNeighborsClassifier(n_neighbors = k, metric = 'minkowski', p = 2)
    classifier.fit(x_train, y_train)
    
    acc_train.append(classifier.score(x_train, y_train))
    acc_test.append(classifier.score(x_test, y_test))
```

```python
from matplotlib import style
style.use('ggplot')
```

```python
k = range(1,21)

plt.figure(figsize = (15,6), dpi = 200)
plt.plot(k, acc_train, '-o', label = 'Training Accuracy')
plt.plot(k, acc_test,  '-o', label = 'Testing Accuracy')

plt.xticks(k)
plt.xlabel('Values of k')
plt.ylabel('Model performance wrt values of k')
plt.legend()
```
